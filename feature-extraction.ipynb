```python
def extract_episodes(txt):
    check = False
    data = ""
    
    for i in txt:
        if i == ")":
            return data
        if check == True:
            data+=i
        if i =="(":
            check = True
```

```python
anime["Title"].apply(extract_episodes)
```

```python
anime["episodes"] = anime["Title"].apply(extract_episodes)
```

```python
anime
```

```python
anime["episodes"] = anime["episodes"].str.replace(" eps","")
```

```python
anime["episodes"] = anime["episodes"].astype(int)
```

```python
def extract_date(sentence):
    return sentence.split(")")[1][:19]
```

```python
anime["date"] = anime["Title"].apply(extract_date)
```
