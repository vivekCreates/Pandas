### Pandas Series

A Series is a one-dimensional labled array capable of holding any data type. This axis labels are collectively called the index.

### Creating Series

There are multiple ways to create a pandas Series

```python
import numpy as np 
import pandas as pd
```

```python
labels = ['a','b','c']
my_list = [1,2,3]
arr = np.array([1,2,3])
d = {1:10,2:20}
```

```python
pd.Series(my_list)
```

```python
pd.Series(my_list,index=labels)
```

```python
pd.Series(arr)
```

```python
pd.Series(d)
```
